```python
SKU_Wrt_ExternOrderNO = pd.merge(CX_SKU_Master,CX_order_report,on = 'SKU',how = 'outer')
SKU_Wrt_ExternOrderNO
```

```python
SKU_Wrt_ExternOrderNO['weight_by_quantity'] = SKU_Wrt_ExternOrderNO['Weight (g)'] * SKU_Wrt_ExternOrderNO['Order Qty']
SKU_Wrt_ExternOrderNO
```

```python
temp1.rename(columns = {'Charged Weight': 'total_wt_cc','Zone' : 'Zone_cc','Billing Amount (Rs.)' : 'charges_by_cc' },inplace = True)
temp1
```

```python
temp1.drop(['d_500'],axis = 1,inplace = True)
temp1.drop(['Warehouse Pincode'], axis = 1,inplace = True)
```

```python
temp1['d_500'] = temp1['total_wt_cc']/500
temp['d_500'] = np.ceil(temp1['d_500'])
temp1['weight_slab_cc'] = temp1['d_500'] * 0.5
temp1
```

```python
temp1.drop(['d_500'],axis = 1,inplace = True)
temp1
```

```python
temp2 = pd.merge(temp1,CX_Pincode_Zones,on = 'Customer Pincode',how = 'outer')
temp2
```

```python
temp2.drop(['Warehouse Pincode'], axis =1,inplace = True)
temp2.drop(['Customer Pincode'], axis =1,inplace = True)
temp2
```

```python
temp2.info()
```

```python
list_zones = temp2['Zone_X'].unique()
list_zones
```

```python
final_df = pd.DataFrame()
final_df
```

```python
value_pair = {'a' : 29.5,'b' : 33,'c' : 40.1,'d' : 45.4,'e' : 56.6}
```

```python
additional_charges = {'a' : 23.6,'b' : 28.3,'c' : 38.9,'d' : 44.8,'e' :55.5}
```

```python
rto_fixed_charges = {'a' : 13.6,'b' : 20.5,'c' : 31.9,'d' : 41.3,'e' : 50.7}
```

```python
rto_additional_charges = {'a' : 23.6,'b' : 28.3,'c' : 38.9,'d' : 44.8,'e' :55.5}
```

```python
for i in list_zones:
    temp3 = temp2[temp2['Zone_X'] == i]
    temp3['fixed'] = value_pair[i]
    temp3['additional'] = np.where(temp3 ['weight_slab_X']*2 > 1,((temp3['weight_slab_X']*2)-1)*additional_charges[i],0)
    temp3['rto_fixed_charges'] =  np.where(temp3['Type of Shipment'].str.contains("RTO", case=False), rto_fixed_charges[i], 0)
    temp3['rto_additional_charges'] =  np.where((temp3 ['weight_slab_X']*2 > 1) & (temp3['Type of Shipment'].str.contains('Forward and RTO charges', case=False)),((temp3['weight_slab_X']*2)-1)*rto_additional_charges[i],0)
    final_df = final_df.append(temp3)
```

```python
final_df.reset_index(inplace = True, drop = True)
final_df
```

```python
final_df['expected_charges_X'] = final_df['fixed'] + final_df['additional'] + final_df['rto_fixed_charges'] + final_df['rto_additional_charges']
final_df
```

```python
final_df['difference_x_cc'] = round(final_df['expected_charges_X'] - final_df['charges_by_cc'],2)
final_df
```

```python
final_df = final_df.drop(columns = ['fixed','additional','rto_fixed_charges','rto_additional_charges'])
final_df
```

```python
DF = final_df[['Order_ID','total_wt_X','weight_slab_X','AWB Code','total_wt_cc','Zone_cc','Type of Shipment','charges_by_cc','weight_slab_cc','Zone_X','expected_charges_X','difference_x_cc']]
DF
```
